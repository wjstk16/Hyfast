```python
import numpy as np
import pandas as pd

df = pd.read_csv('./10feature_cpu.csv')
df_time = pd.read_csv('./csv/10feature_cpu_date.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

col = list(map(str, df.columns))
x = df[col[:-1]]
y = df[col[-1]]

scaler = StandardScaler()
x = scaler.fit_transform(x)
x = pd.DataFrame(x)
x.columns = col[:-1]


#print(df_scaled)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4)
x_test, x_val, y_test, y_val = train_test_split(x_test,y_test, test_size=0.5)
```

```python
import tensorflow as tf
import keras

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import multi_gpu_model
#tf.compat.v1.disable_eager_execution()

x_train=x_train.astype(float)
y_train=y_train.astype(float)
with tf.device('/gpu:1'):
    model = Sequential()
    model.add(Dense(10,activation='relu', input_dim= x_train.shape[1]))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(10,activation='relu'))
    #model.add(Dense(400,activation='relu'))
    #model.add(Dense(400,activation='relu'))
    #model.add(Dense(400,activation='relu'))
    #model.add(Dense(500,activation='relu'))
    #model.add(Dense(600,activation='relu'))
    model.add(Dense(1))
    #model = multi_gpu_model(model,gpus=4)
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    model.fit(x_train, y_train, epochs=5000, batch_size=2048)

    #Y_pred = model.predict_classes(x_val)
    Y_pred = (model.predict(x_val) > 0.5).astype("int32")
    
#y_val.to_numpy()
#print('Misclassified Samples:', (y_val != Y_pred).sum())
    Y_pred = np.array(np.array(Y_pred).flatten().tolist())
    print('Misclassified Samples:', (y_val != Y_pred).sum())
    print('accuracy:', (1 - (y_val != Y_pred).sum()/len(y_val))*100)
```
